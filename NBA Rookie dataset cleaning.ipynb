```python
import pandas as pd
from unidecode import unidecode

# Replace 'your_file_path' with the actual path where the file is located
file_path = "rookies_by_year (1).csv"

# Load the CSV file
data = pd.read_csv(file_path)

# Normalize player names in the dataset
data['Player'] = data['Player'].apply(lambda x: unidecode(x))

# Display the first few rows
data.head()
```

Removing every entry before the NBA merger in 1976

```python
# Ensure the 'Season' column is treated as numeric
data['Season'] = data['Season'].astype(str).str[:4].astype(int)

# Filter the dataset to include only the 1976-1977 season and beyond
data_filtered = data[data['Season'] >= 1976]

# Display the last five rows of the updated filtered DataFrame
data_filtered.tail()
```

Making TOV in to TOV per Game and removing unnecessary stats

```python
# Update the list of columns to retain, accounting for exact column names
columns_to_keep = [
    "Player", "Age", "G", "MP", "FG%", "3P%", "FT%", 
    "PER GAME MP", "PER GAME  PTS", "PER GAME TRB", 
    "PER GAME AST", "PER GAME STL", "PER GAME BLK"
]

# Filter the dataset to retain the specified columns
filtered_dataset = data[columns_to_keep]

# Calculate PER GAME TOV (Turnovers Per Game)
filtered_dataset["PER GAME TOV"] = data["TOV"] / data["G"]


# Show a preview of the filtered dataset
filtered_dataset.head()
```

```python
# Round the PER GAME TOV column to 1 decimal place
filtered_dataset["PER GAME TOV"] = filtered_dataset["PER GAME TOV"].round(1)


# Show a preview of the updated dataset
filtered_dataset.head()
```
